```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('Marketing Team Data - Cleaned data.csv')

# Convert 'Amount Spent in INR' to numerical (remove $ and commas if necessary)
df['Amount Spent in INR'] = df['Amount Spent in INR'].replace('[\$,]', '', regex=True).astype(float)

# Convert other financial metrics like CPC, CPR to numerical
df['Cost Per Click (CPC)'] = df['Cost Per Click (CPC)'].replace('[\$,]', '', regex=True).astype(float)
df['Cost per Result (CPR)'] = df['Cost per Result (CPR)'].replace('[\$,]', '', regex=True).astype(float)

# Convert 'Click-Through Rate (CTR)' to percentage values (remove %)
df['Click-Through Rate (CTR)'] = df['Click-Through Rate (CTR)'].str.rstrip('%').astype('float')

# Visual 1: Bar plot - CTR of Different Campaigns
plt.figure(figsize=(12, 8))
sns.barplot(x='Campaign Name', y='Click-Through Rate (CTR)', data=df, ci=None, palette="viridis")
plt.title('Click-Through Rate (CTR) of Different Campaigns')
plt.xticks(rotation=90)
plt.show()

# Visual 2: Box plot - CPC across Campaigns
plt.figure(figsize=(12, 8))
sns.boxplot(x='Campaign Name', y='Cost Per Click (CPC)', data=df)
plt.title('Cost Per Click (CPC) by Campaign')
plt.xticks(rotation=90)
plt.show()

# Visual 3: Scatter plot - Reach vs. Amount Spent in INR
plt.figure(figsize=(12, 8))
sns.scatterplot(x='Reach', y='Amount Spent in INR', hue='Campaign Name', data=df, s=100)
plt.title('Reach vs. Amount Spent (INR) by Campaign')
plt.show()

# Visual 4: Line plot - Amount Spent over Time (if you have date/time information)
# If no time column exists, this is just a placeholder. Otherwise, replace 'Date Column' with the correct column.
# plt.figure(figsize=(12, 8))
# sns.lineplot(x='Date Column', y='Amount Spent in INR', hue='Campaign Name', data=df)
# plt.title('Amount Spent Over Time')
# plt.xticks(rotation=90)
# plt.show()

# Visual 5: Histogram - Distribution of CTR
plt.figure(figsize=(12, 8))
plt.hist(df['Click-Through Rate (CTR)'], bins=10, color='blue', alpha=0.7)
plt.title('Distribution of Click-Through Rate (CTR)')
plt.xlabel('Click-Through Rate (CTR)')
plt.ylabel('Frequency')
plt.show()

# Visual 6: Heatmap - Correlation between numerical variables
plt.figure(figsize=(12, 8))
correlation = df[['Reach', 'Impressions', 'Clicks', 'Unique Clicks', 'Amount Spent in INR', 'Cost Per Click (CPC)', 'Cost per Result (CPR)', 'Click-Through Rate (CTR)']].corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm', linewidths=1)
plt.title('Correlation Heatmap of Metrics')
plt.show()

# Visual 7: Pie chart - Total Amount Spent per Campaign
plt.figure(figsize=(10, 8))
df.groupby('Campaign Name')['Amount Spent in INR'].sum().plot(kind='pie', autopct='%1.1f%%', colors=sns.color_palette("pastel"))
plt.title('Total Amount Spent per Campaign')
plt.ylabel('')
plt.show()

# Visual 8: Bar plot - Impressions per Campaign
plt.figure(figsize=(12, 8))
sns.barplot(x='Campaign Name', y='Impressions', data=df, ci=None, palette="magma")
plt.title('Total Impressions per Campaign')
plt.xticks(rotation=90)
plt.show()

# Visual 9: Box plot - Performance of Different Audience Groups (Audience vs CTR)
plt.figure(figsize=(12, 8))
sns.boxplot(x='Audience', y='Click-Through Rate (CTR)', data=df)
plt.title('CTR by Audience Group')
plt.xticks(rotation=45)
plt.show()

# Visual 10: Line plot - Frequency vs. Clicks by Campaign (to analyze ad saturation over campaigns)
plt.figure(figsize=(12, 8))
sns.lineplot(x='Frequency', y='Clicks', hue='Campaign Name', data=df)
plt.title('Frequency vs Clicks by Campaign')
plt.show()

# Performance Evaluation
df['Performance Score'] = df['Click-Through Rate (CTR)'] - (df['Cost Per Click (CPC)'] + df['Cost per Result (CPR)'])
poor_performance = df.groupby('Campaign Name')['Performance Score'].mean().sort_values().head(1)

print("Ad to discontinue based on poor performance:", poor_performance)
```

```python
import pandas as pd

# Load the dataset
df = pd.read_csv('Marketing Team Data - Cleaned data.csv')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Bar plot for Audience Distribution by Campaign
plt.figure(figsize=(10,6))
sns.countplot(x='Campaign Name', hue='Audience', data=df)
plt.title('Audience Distribution by Campaign')
plt.xticks(rotation=90)
plt.show()
```

```python
# Bar plot for Age Group Distribution by Campaign
plt.figure(figsize=(10,6))
sns.countplot(x='Campaign Name', hue='Age', data=df)
plt.title('Age Group Distribution by Campaign')
plt.xticks(rotation=90)
plt.show()
```

```python
# Box plot to analyze CTR performance by Audience
plt.figure(figsize=(10,6))
sns.boxplot(x='Audience', y='Click-Through Rate (CTR)', data=df)
plt.title('CTR Performance by Audience')
plt.xticks(rotation=90)
plt.show()

# Box plot to analyze CPC performance by Age group
plt.figure(figsize=(10,6))
sns.boxplot(x='Age', y='Cost Per Click (CPC)', data=df)
plt.title('CPC Performance by Age Group')
plt.xticks(rotation=90)
plt.show()
```

```python
# Create a performance score based on CTR and CPC (this is just an example)
# Convert the relevant columns to numeric, handling errors by coercing to NaN
df['Click-Through Rate (CTR)'] = pd.to_numeric(df['Click-Through Rate (CTR)'], errors='coerce')
df['Cost Per Click (CPC)'] = pd.to_numeric(df['Cost Per Click (CPC)'], errors='coerce')

# Now calculate the performance score
df['Performance Score'] = df['Click-Through Rate (CTR)'] - df['Cost Per Click (CPC)']

# Sort campaigns by performance score
worst_performers = df[['Campaign Name', 'Performance Score']].groupby('Campaign Name').mean().sort_values(by='Performance Score').head(5)

print("Top campaigns to consider for discontinuation based on performance score:")
print(worst_performers)
```
